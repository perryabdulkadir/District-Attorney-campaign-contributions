```python
pw = ['Police', 'police', 'Law Enforcement', 'law enforcement', 'Officer', 'officer', 'Public Safety', 'public safety', 
 'Sheriff', 'sheriff', 'Correction', 'correction', 'offs', 'Offs', 'Corr.', 'corr.', 'Captain', 'captain', 
 'detective', 'Detective', 'Prison', 'prison', 'jail', 'Jail']
sf_da_df['Tran_Occ_Emp'] = sf_da_df["Tran_Occ"] + ' ' + sf_da_df["Tran_Emp"]
```

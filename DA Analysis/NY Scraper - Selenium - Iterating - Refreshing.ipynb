```python
##nested for loop

candidate_list = []

# for opt in candidate_committee.options:
#     print(opt.text)
#     opt.click()
#     candidate_list.append(opt.text)
#     #set max campaign contribution to $100 mil 100000000.00 
#     sleep(5)

for i in range(518):
    try:
        #select "County" from Filer Type dropdown 
        # #navigate to NY elections site
        browser.get('https://publicreporting.elections.ny.gov/Contributions/Contributions')
        #select "County" from Filer Type dropdown 
        sleep(3)
        filer_type = Select(browser.find_element_by_name('lstUCOfficeType'))
        filer_type.select_by_visible_text('County')
        sleep(3)
        #select "All" from County dropdown
        county = Select(browser.find_element_by_name('lstUCCounty'))
        county.select_by_visible_text('All')
        sleep(3)
        #select "Office" from Search By dropdown
        search_by = Select(browser.find_element_by_name('ddlSearchBy'))
        search_by.select_by_visible_text('Office')
        sleep(3)
        #select "Office" from Search By dropdown
        office = Select(browser.find_element_by_name('lstUCOffice')) 
        office.select_by_visible_text('District Attorney')
        sleep(3)
        #select "All" from District dropdown
        district = Select(browser.find_element_by_name('lstUCDistrict')) 
        district.select_by_visible_text('All')
        sleep(3)    
        #select "X" from Candidate/Committee name
        candidate_committee = Select(browser.find_element_by_name('ddlFiler'))
        for j, opt in enumerate(candidate_committee.options):
            if j == i:
                opt.click()
                sleep(7)
#         amount_max_dollars = browser.find_element_by_name('txtExpenditureAmountMax')
#         amount_max_dollars.send_keys('100000000.00')
#         sleep(3)
        #select "Date Range" from Select Date dropdown
        select_date = Select(browser.find_element_by_name('ddlSelectDate')) 
        select_date.select_by_visible_text('Date Range')
        sleep(7)
        #Enter "01/01/2000" to the Date From box
        date_from = browser.find_element_by_name('txtDateFrom')
        date_from.send_keys('01/01/2000')
        sleep(7)
        #Enter "03/15/2021" to the Date To box
        date_to = browser.find_element_by_name('txtDateTo')
        date_to.send_keys('03/15/2020')
        sleep(10)
        #click search button
        search = browser.find_element_by_name('Command')
        search.click()
        sleep(10)
        #click CSV download
        csv_download = browser.find_element_by_id('btnCSVDownloadData')
        csv_download.click()
        sleep(10)
        browser.refresh()
    except:
        print('No data for this candidate')
        browser.refresh()
```
